```python
df_data = df_data.reset_index()

df_data['TRADE_CODE'] =df_data['index']
df_data = df_data.drop(['index'],axis=1)
df_data
df_data.columns = ['bond code','bond name','stock code','stock name',
                    'stock price','convert price','bond life','premium rate','bond balance']
df_data
```

```python
df_data = df_data.dropna(axis=0,how='any')
df_data['moneyness'] = (df_data['stock price'] - df_data['convert price']) /  df_data['stock price']
df_data['balance/life'] = (df_data['bond balance'])/(df_data['bond life'])
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
tmp_score =  df_data['balance/life'].values.reshape(-1,1)
score_minmax_normalized = min_max_scaler.fit_transform(tmp_score)
score_minmax_normalized
df_data['balance/life minmax normalized'] = score_minmax_normalized


tmp_premium_rate = df_data['premium rate'].values.reshape(-1,1)
premium_rate_minmax_normalized = min_max_scaler.fit_transform(tmp_premium_rate)
df_data['premium rate minmax normalized'] = premium_rate_minmax_normalized
```

# $MinMaxScale(\frac{bond\  balance}{bond\ life}) + moneyness - MinMaxScale(premium\ rate)$

```python
df_data['score final'] = df_data['balance/life minmax normalized'] + df_data['moneyness'] - df_data['premium rate minmax normalized']
```
