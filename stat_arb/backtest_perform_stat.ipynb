```python
import os
import pandas as pd
path= '/Users/bokie/code/python/CMS/可转债套利/cms/stat_arb/pair_perform'
files = os.listdir(path)
dfs = []
for file in files:
    tmp_df = pd.read_excel(f'{path}/{file}')
    dfs.append(tmp_df)
df_peform_total = pd.concat(dfs,axis=0)    
pd.set_option('display.float_format', lambda x: '%.3f' % x)
df_peform_total.sort_values(by='net profit',ascending=False)
```
