```python
stk_codes = df_stk_clean['code'].tolist()

df_stk_prices = []
for stk_code in stk_codes[0:1]+stk_codes[3:5]:

    df_tmp = w.wsi(f"{stk_code}", "close", "2022-09-01 09:00:00", "2022-11-10 15:00:00", "",usedf=True)[1]
    df_tmp.columns = [stk_code]
    df_tmp[stk_code] = np.log(df_tmp[stk_code])
    # df_stk_price2 = w.wsi("600798.SH", "close", "2022-09-01 09:00:00", "2022-11-10 15:00:00", "",usedf=True)[1]
    # df_stk_price2.columns = ["600798.SH"]
    df_stk_prices.append(df_tmp)
```

```python
df_stk_price = pd.concat(df_stk_prices,axis=1)
df_stk_price.corr(method='spearman')
```

# stock cointegration

```python
import statsmodels.api as sm
import statsmodels.tsa.stattools as ts

# y = np.random.normal(0,1, 250)
# x = np.random.normal(0,1, 250)

y = df_stk_price['601006.SH']
x = df_stk_price['002027.SZ']

def cointegration_test(y, x):
    # Step 1: regress on variable on the other 
    ols_result = sm.OLS(y, x).fit() 
    # Step 2: obtain the residual (ols_resuld.resid)
    # Step 3: apply Augmented Dickey-Fuller test to see whether 
    #        the residual is unit root    
    return ts.adfuller(ols_result.resid)
cointegration_test(y, x)[1]    
```
